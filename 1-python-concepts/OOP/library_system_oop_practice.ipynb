```python
# Build a Library system with these classes:

# 1. Book
#    - __init__: title, author, pages
#    - __str__: "The Hobbit by J.R.R Tolkien (300 pages)"
#    - __repr__: "Book('The Hobbit', 'Tolkien', 300)"
#    - @property: summary
#      returns "Title: X | Author: Y | Pages: Z"

# 2. Library — inherits nothing, standalone class
#    - class variable: total_libraries = 0
#      increments every time a Library is created
#    - __init__: name, city
#    - add_book(book): adds to internal list
#    - remove_book(title): removes by title,
#      print "Book not found" if missing
#    - find_book(title): returns the Book object
#      or None if not found
#    - __len__: returns number of books
#    - __str__: "City Library — 10 books"

# 3. DigitalLibrary — inherits from Library
#    - extra attribute: max_books = 5
#      (digital library has a limit)
#    - override add_book() to enforce the limit
#      print "Library full" if limit reached
#    - classmethod: from_string(cls, s)
#      s = "TechLib-Kathmandu"
#      creates DigitalLibrary("TechLib", "Kathmandu")
```

```python
class Book:

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author} ({self.pages} pages)"
    
    def __repr__(self):
        return "Book({}, {}, {})".format(self.title, self.author, self.pages)
    
    @property
    def summary(self):
        return "Title: {} | Author: {} | Pages: {}".format(self.title, self.author, self.pages)
    

class Library:
    total_libraries = 0
    def __init__(self, name, city):
        self.name = name
        self.city = city
        self.books = []
        Library.total_libraries += 1

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, title):
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                return
        print("Book not found")

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                return book
        return None
    
    def __len__(self):
        return len(self.books)
    
    def __str__(self):
        return "{} Library - {} books".format(self.name, len(self.books))
    

class DigitalLibrary(Library):
    max_books = 5

    def __init__(self, name, city):
        super().__init__(name, city)

    def add_book(self, book):
        if len(self.books) >= self.max_books:
            print("Library full")
        else:
            self.books.append(book)

    @classmethod
    def from_string(cls, s):
        name, city = s.split("-")
        return cls(name, city)
```

```python
## Test it
b1 = Book("The Hobbit", "Tolkien", 300)
b2 = Book("Clean Code", "Martin", 464)
b3 = Book("Python Crash Course", "Matthes", 544)
b4 = Book("Atomic Habits", "James Clear", 320)
b5 = Book("Deep Work", "Cal Newport", 296)
b6 = Book("The Pragmatic Programmer", "Andrew Hunt", 352)

lib = Library("City Library", "Kathmandu")
lib.add_book(b1)
lib.add_book(b2)

print(lib)                        # City Library — 2 books
print(len(lib))                   # 2
print(lib.find_book("Clean Code"))  # Clean Code by Martin (464 pages)

dlib = DigitalLibrary.from_string("TechLib-Kathmandu")
dlib.add_book(b1)
dlib.add_book(b2)
dlib.add_book(b3)
dlib.add_book(b4)
dlib.add_book(b5)
dlib.add_book(b6)   # Library full

print(dlib)         # TechLib — 5 books
print(len(dlib))    # 5
```
