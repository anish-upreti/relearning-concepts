## What is Exception Handling?
Exception handling is Python's way of dealing with errors gracefully instead of letting your program crash. When something goes wrong (like dividing by zero or opening a file that doesn't exist), Python raises an exception. Exception handling lets you catch these errors and respond to them.


````python
# basic structure of try and except
try:
    # Code that might cause an error
    result = 10 / 0
except:
    # What to do if an error occurs
    print("Oops! Something went wrong")
````

````python
## some common error types
# ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")

# ValueError (wrong type of value)
try:
    number = int("hello")
except ValueError:
    print("That's not a valid number!")

# FileNotFoundError
try:
    file = open("nonexistent.txt")
except FileNotFoundError:
    print("File not found!")

# KeyError (dictionary key doesn't exist)
try:
    my_dict = {"name": "Alice"}
    print(my_dict["age"])
except KeyError:
    print("That key doesn't exist!")
````

````python
## calling multiple exceptions
try:
    age = int(input("Enter your age: "))
    result = 100 / age
except ValueError:
    print("Please enter a number!")
except ZeroDivisionError:
    print("Age cannot be zero!")
````

````python
# one block for multiple exceptions
try:
    age = int(input("Enter your age: "))
    result = 100 / age
except (ValueError, ZeroDivisionError) as e:
    print(f"Error: {e}")
````

````python
try:
    # Try to do something
    number = int(input("Enter a number: "))
    result = 10 / number
except ValueError:
    # Runs if ValueError occurs
    print("Invalid input!")
except ZeroDivisionError:
    # Runs if ZeroDivisionError occurs
    print("Cannot divide by zero!")
else:
    # Runs ONLY if no exception occurred
    print(f"Result is: {result}")
finally:
    # ALWAYS runs, whether there was an error or not
    print("Execution completed")
````

````python
## getting error information
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error type: {type(e).__name__}")
    print(f"Error message: {e}")
````

````python
## raising our own exceptions
def check_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative!")
    if age > 150:
        raise ValueError("Age seems unrealistic!")
    return True

try:
    check_age(200)
except ValueError as e:
    print(f"Invalid age: {e}")
````

````python
# ============================================
#  Student Grade Management System
# ============================================

# Create 3 custom exceptions:
# 1. GradeOutOfRangeError - when grade not in 0-100
# 2. DuplicateCourseError - when course already exists
# 3. CourseNotFoundError - when course doesn't exist

# Build Student class:

# __init__(name, student_id)
# - Initialize name and student_id
# - Create empty _grades dictionary
# - _grades stores {course_name: grade}

# Implement methods:

# add_course(course_name, grade):
# - Raise DuplicateCourseError if course exists
#   (show existing grade in message)
# - Raise GradeOutOfRangeError if grade not in 0-100
# - Add course if valid

# update_grade(course_name, new_grade):
# - Raise CourseNotFoundError if course doesn't exist
#   (list available courses)
# - Raise GradeOutOfRangeError if grade not in 0-100
# - Return (old_grade, new_grade)

# get_gpa():
# - Calculate GPA on 4.0 scale:
#     90-100 = 4.0
#     80-89  = 3.0
#     70-79  = 2.0
#     60-69  = 1.0
#     <60    = 0.0
# - Return 0.0 if no courses
# - Round to 2 decimals

# get_transcript():
# - Return a copy of grades dictionary

# __str__():
# - Return "Student(name, ID: student_id, GPA: gpa)"

# Write tests:
# - Add 3 courses
# - Try duplicate (catch exception)
# - Try invalid grade 150 (catch exception)
# - Try update non-existent course (catch exception)
# - Successfully update a grade
# - Print student info and transcript
````

````python
# creating custom exceptions
class GradeOutOfRangeError(Exception):
    pass

class DuplicateCourseError(Exception):
    pass

class CourseNotFoundError(Exception):
    pass

# defining student class
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

        # Private dictionary to store course grades
        # Format: {course_name: grade}
        self._grades = {}

    def add_course(self, course_name, grade):
        # checking if course already exists
        if course_name in self._grades:
            raise DuplicateCourseError(f"Course {course_name} already exists with grades {self._grades[course_name]}")
        
        # if invalid grade
        if grade < 0 or grade > 100:
            raise GradeOutOfRangeError(f"Grade {grade} is out of Range. It should be between 0 and 100.")
        
        self._grades[course_name] = grade

    def update_grade(self, course_name, new_grade):
        if course_name not in self._grades:
            raise CourseNotFoundError(f"Course {course_name} does not exist. The available courses are: {self._grades}")
        
        if new_grade < 0 or new_grade > 100:
            raise GradeOutOfRangeError(f"Grade {new_grade} is out of range.")
        
        # store old grade before overriding
        old_grade = self._grades[course_name]

        # new grade updation
        self._grades[course_name] = new_grade

        return (old_grade, new_grade)
    
    def get_gpa(self):
        # return 0.0 if student has no courses
        if not self._grades:
            return 0.0
        
        # list to store converted grade points
        grade_points = []

        # convert each numeric grade to gpa scale
        for grade in self._grades.values():
            if 90 <= grade <= 100:
                grade_points.append(4.0)
            elif 80 <= grade <= 89:
                grade_points.append(3.0)
            elif 70 <= grade <= 79:
                grade_points.append(2.0)
            elif 60 <= grade <= 69:
                grade_points.append(1.0)
            else:
                grade_points.append(0.0)

        # calculate averge gpa and round to 2 decimal places
        average_gpa = round(sum(grade_points) / len(grade_points), 2)

        return average_gpa
    
    def get_transcript(self):
        # return a copyt of grades dictionary
        return self._grades.copy()
    
    def __str__(self):
        return f"Student({self.name}, ID: {self.student_id}, GPA = {self.get_gpa()})"
        
````

````python
# ============================================
# Tests
# ============================================

# Create a student object
student = Student("Alice", "S101")

# Add 3 valid courses
student.add_course("Math", 95)
student.add_course("Science", 82)
student.add_course("History", 76)

# Try adding a duplicate course
try:
    student.add_course("Math", 88)
except DuplicateCourseError as e:
    print("DuplicateCourseError caught:", e)

# Try adding an invalid grade
try:
    student.add_course("English", 150)
except GradeOutOfRangeError as e:
    print("GradeOutOfRangeError caught:", e)

# Try updating a non-existent course
try:
    student.update_grade("Geography", 85)
except CourseNotFoundError as e:
    print("CourseNotFoundError caught:", e)

# Successfully update an existing course grade
old_grade, new_grade = student.update_grade("Science", 89)
print(f"Updated Science grade from {old_grade} to {new_grade}")

# Print student information
print(student)

# Print transcript
print("Transcript:", student.get_transcript())
````

## Custom Exceptions & Usage in Classes

## What are Custom Exceptions?

Custom exceptions are user-defined error types that help make code more readable, specific, and maintainable.
```python
class GradeOutOfRangeError(Exception):
    pass
```

- We inherit from `Exception`
- `pass` means no extra logic is added
- Still fully functional as an error type

## Why Inherit from Exception?
```python
class MyError(Exception):
    pass
```

- Python only allows raising objects derived from `BaseException`
- `Exception` is the standard parent class for custom errors
- Inheritance gives built-in behavior like:
  - Storing messages
  - Being caught with `try/except`

## Why is `pass` Enough?
```python
class GradeOutOfRangeError(Exception):
    pass
```

- The parent (`Exception`) already handles everything
- We only need a new error type (name)
- Use custom `__init__` only if extra behavior is needed

## Defining vs Using Exceptions

**Defining (Inheritance happens here):**
```python
class DuplicateCourseError(Exception):
    pass
```

**Using (No inheritance here):**
```python
raise DuplicateCourseError("Course already exists")
```

- Here we are raising, not defining
- We are creating an object of the exception class

## Why Student Class Does Not Inherit from Exceptions
```python
class Student:
    def add_course(self, course_name, grade):
        if course_name in self._grades:
            raise DuplicateCourseError("Already exists")
```

- `Student` is a normal class (not an error)
- It uses exceptions, it does not inherit from them

**Incorrect:**
```python
class Student(DuplicateCourseError):  # Wrong
    pass
```

**Why:**
- A `Student` is NOT an error
- It only raises errors when something goes wrong

## What Does `raise` Actually Do?
```python
raise GradeOutOfRangeError("Invalid grade")
```

**Equivalent to:**
```python
error = GradeOutOfRangeError("Invalid grade")
raise error
```

**What happens:**
- Creates an exception object
- Stops execution
- Passes control to nearest `except` block

# The .copy() Method

## What It Does

`.copy()` creates a separate copy of a dictionary, list, or set.
```python
new_dict = original_dict.copy()
new_list = original_list.copy()
```

## Why Use It?

Without `.copy()`, assignment creates a reference to the same object:
```python
original = {"Math": 90, "Science": 85}
referenced = original  # Both point to the same dictionary

referenced["Math"] = 50
print(original)  # {'Math': 50, 'Science': 85} - changed!
```

With `.copy()`, you get a separate object:
```python
original = {"Math": 90, "Science": 85}
copied = original.copy()  # New dictionary created

copied["Math"] = 50
print(original)  # {'Math': 90, 'Science': 85} - unchanged
print(copied)    # {'Math': 50, 'Science': 85}
```

## Use in Classes

When returning internal data from a class method:
```python
class Student:
    def __init__(self):
        self._grades = {}
    
    def get_transcript(self):
        return self._grades.copy()  # Return a copy, not the original
```

**Why?**
- Protects internal data from outside modification
- Maintains encapsulation
- Prevents accidental changes

**Without .copy():**
```python
transcript = student.get_transcript()
transcript["Math"] = 0  # Modifies student's actual grades!
```

**With .copy():**
```python
transcript = student.get_transcript()
transcript["Math"] = 0  # Only modifies the copy
```

## Key Points

- `=` does NOT make a new object (creates a reference)
- `.copy()` DOES make a new object (creates a copy)
- Use `.copy()` to prevent unintended modifications
- Works with dictionaries, lists, and sets

## Shallow Copy vs Deep Copy

### Shallow Copy (.copy())

`.copy()` creates a shallow copy - only copies the outer container, nested objects are still referenced:
```python
original = {"student": {"name": "Alice", "age": 20}}
copied = original.copy()

copied["student"]["name"] = "Bob"
print(original)  # {"student": {"name": "Bob", "age": 20}} - nested object changed!
```

**What happened?**
- The outer dictionary was copied
- The inner dictionary is still the same object

### Deep Copy (copy.deepcopy())

`deepcopy()` creates a deep copy - copies everything, including nested objects:
```python
import copy

original = {"student": {"name": "Alice", "age": 20}}
deep_copied = copy.deepcopy(original)

deep_copied["student"]["name"] = "Bob"
print(original)      # {"student": {"name": "Alice", "age": 20}} - unchanged!
print(deep_copied)   # {"student": {"name": "Bob", "age": 20}}
```

### Visual Comparison
```python
import copy

original = {
    "course": "Python",
    "students": ["Alice", "Bob"]
}

# Shallow copy
shallow = original.copy()
shallow["students"].append("Charlie")
print(original["students"])  # ['Alice', 'Bob', 'Charlie'] - changed!

# Deep copy
deep = copy.deepcopy(original)
deep["students"].append("David")
print(original["students"])  # ['Alice', 'Bob', 'Charlie'] - unchanged!
```

### When to Use Each

**Use `.copy()` (shallow) when:**
- Simple data structures (no nesting)
- Nested objects don't need to be independent
- Performance matters (faster than deepcopy)

**Use `copy.deepcopy()` when:**
- Nested data structures (lists in dicts, dicts in lists, etc.)
- You need complete independence from original
- Modifying nested objects should not affect original

### Complete Example
```python
import copy

class Student:
    def __init__(self, name):
        self.name = name
        self._courses = {
            "Math": {"grade": 90, "credits": 3},
            "Science": {"grade": 85, "credits": 4}
        }
    
    def get_transcript_shallow(self):
        return self._courses.copy()
    
    def get_transcript_deep(self):
        return copy.deepcopy(self._courses)

student = Student("Alice")

# Shallow copy problem
transcript1 = student.get_transcript_shallow()
transcript1["Math"]["grade"] = 0
print(student._courses["Math"]["grade"])  # 0 - original changed!

# Deep copy solution
transcript2 = student.get_transcript_deep()
transcript2["Science"]["grade"] = 0
print(student._courses["Science"]["grade"])  # 85 - original safe!
```
