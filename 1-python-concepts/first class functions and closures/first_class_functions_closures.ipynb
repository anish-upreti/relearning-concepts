## First-Class Functions

---

## What are First-Class Functions?

In Python, functions are **first-class citizens** (or **first-class objects**).

This means functions can be treated just like any other object (strings, integers, lists, etc.).

---

## What Can You Do with First-Class Functions?

### 1. **Assign functions to variables**
### 2. **Pass functions as arguments to other functions**
### 3. **Return functions from other functions**
### 4. **Store functions in data structures** (lists, dictionaries, etc.)

---


````python
## Assign functions to variables
# Functions can be assigned to variables like any other value.

def greet():
    return "Hello"

say_hello = greet  # we are not using parenthesis because we are assigning it and not calling it

# Call using the new variable
print(say_hello())  # Output: Hello!
print(greet())      # Output: Hello!

````

````python
### Without Parentheses vs With Parentheses

def greet():
    return "hello"

print(greet)   # function object
print(greet())    # function call

# - `greet` → the function itself (object)
# - `greet()` → calls the function and returns the result
````

````python
## 2. Pass Functions as Arguments

# You can pass a function as an argument to another function.

def greet():
    return "Hello!"

def shout():
    return "YELLING!"

def call_func(func):
    """Takes a function and calls it"""
    result = func()  # Call the passed function
    print(result)

call_func(greet)  # Output: Hello!
call_func(shout)  # Output: YELLING!
````

````python
### Real Example: `map()` function
def square(x):
    return x * x

nums = [1,2,3,4,5]

squares = map(square, nums)
print(squares)
print(list(squares))


# `map()` takes a function (`square`) and applies it to each element.
````

````python
### Another Example: Custom Sorting

def get_length(word):
    return len(word)

words = ["apple", "pie", "zoo", "elephant"]

# Sort by length using a function
sorted_words = sorted(words, key=get_length)   # key parameter is accepting a function
print(sorted_words)
````

````python
## 3. Return Functions from Other Functions

def outer_func():
    def inner_func():
        return "Hello from inner function"
    return inner_func

# Get the inner function
my_func = outer_func()

# Now call it
print(my_func())


# 1. `outer()` is called → returns `inner` function
# 2. `my_func` now holds the `inner` function
# 3. `my_func()` calls the `inner` function
````

````python
### Practical Example: Function Factory

def make_multiplier(n):
    def multiplier(x):
        return n * x
    return multiplier

times_3 = make_multiplier(3)
times_5 = make_multiplier(5)

print(times_3(10))  # Output: 30
print(times_5(10))  # Output: 50

# - `make_multiplier(3)` returns a function that multiplies by 3
# - `make_multiplier(5)` returns a function that multiplies by 5
````

````python
## 4. Store Functions in Data Structures
# Functions can be stored in lists, dictionaries, sets, etc.

### Store in a list
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

# store functions in a list
operations = [add, subtract, multiply]

# using the operations
print(operations[0](1, 2))
print(operations[1](4, 3))
print(operations[2](2, 3))


## Storing the functions in a dictionary
calculator = {
    'add': add,
    'subtract': subtract,
    'multiply': multiply
}

# using the function in a dictionary
print(calculator["add"](10, 20))
print(calculator["multiply"](1, 3))
````

## Why do first class functions matter?
First-class functions enable powerful programming patterns:  
- Callbacks
- higher order functions
- decorators  
etc

````python
# callbacks - A callback is a function that is passed as an argument to another function and is executed later.

def process_data(data, callback):
    result = data * 2
    callback(result)

def print_result(value):
    print(f"Result: {value}")

process_data(5, print_result)
````

````python
### 2. Higher-Order Functions

# Functions that take other functions as arguments or return functions.

# Filter even numbers
nums = [1, 2, 3, 4, 5, 6, 7, 8]

def is_even(n):
    # return n%2 == 0
    if n%2 == 0:
        return n

evens = filter(is_even, nums)
print(evens)
print(list(evens))
````

````python
### 3. **Decorators**

# Decorators are possible because functions are first-class! A decorator is a function that modifies another function’s behavior
# without changing its code.“Wrap a function with extra functionality”

def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@my_decorator
def say_hello():
    print("hello")

say_hello()     # it is same as say_hello = my_decorator(say_hello)
# say_hello() actually runs the wrapper()
````

## Python Closures
---

## What is a Closure?

A **closure** is a function that:
1. Is defined inside another function (nested function)
2. **Remembers** variables from its enclosing scope
3. Can access those variables even after the outer function has finished executing

Think of it as a **function with memory** 🧠.

````python

## Simple Example
def outer(x):
    def inner():
        print(x)  # inner() remembers 'x'
    return inner

my_func = outer(10)
my_func()
````

**Step 1:** `outer(10)` is called
- `x = 10` in outer's scope
- `inner()` is defined (but not called)
- `outer()` returns the `inner` function

**Step 2:** `outer()` finishes and its scope normally should be destroyed
- BUT: `inner()` still needs `x`
- Python keeps `x` alive for `inner()`

**Step 3:** `my_func()` is called
- `inner()` executes
- It accesses `x = 10` (from the closed-over scope)
- Prints `10`

---

## Why Do We Need Closures?


````python
## Data encapsulation (hide data)
def make_counter():
    count = 0   # private variable

    def increment():
        nonlocal count
        count += 1
        return count
    return increment

counter = make_counter()
print(counter())  # 1
print(counter())  # 2
print(counter())  # 3

# Key Point: `count` is private. You can't access it directly, only through `increment()`.
````

````python
## 3 Callback Functions with State
def create_button_handler(button_id):
    def on_click():
        print(f"Button {button_id} clicked!")
    return on_click

button1_handler = create_button_handler(1)
button2_handler = create_button_handler(2)

button1_handler()  # Button 1 clicked!
button2_handler()  # Button 2 clicked!

# Each handler remembers its own `button_id`.
````

## The `nonlocal` Keyword

When you want to **modify** a variable from the outer scope, use `nonlocal`.

### Without `nonlocal` (Error)
```python
def outer():
    count = 0
    
    def inner():
        count = count + 1  # ❌ Error: UnboundLocalError
        return count
    
    return inner
```

**Why?** Python thinks you're creating a **new local** variable `count`.

---

### With `nonlocal` (Correct)
```python
def outer():
    count = 0
    
    def inner():
        nonlocal count  # Tell Python to use outer's count
        count = count + 1
        return count
    
    return inner

counter = outer()
print(counter())  # 1
print(counter())  # 2
```

## Closures vs Regular Functions

---

## Part 1: Core Difference

### 🟢 Regular Function
```python
x = 10

def func():
    print(x)

func()  # 10
x = 20
func()  # 20
```

**What's happening?**
* `func` does **NOT** store `x`
* It just **looks up** `x` when it runs
* So it always uses the **latest value**

👉 **It depends on external variable**

---

### 🔵 Closure
```python
def outer(x):
    def inner():
        print(x)
    return inner

func1 = outer(10)
func2 = outer(20)

func1()  # 10
func2()  # 20
```

**What's happening?**
* `inner` **remembers** `x`
* Each function has its **own stored value**

👉 **It captures value at creation time**

---

## ⚡ Key Difference (Simple Table)

| Feature | Regular Function | Closure |
|---------|------------------|---------|
| **Variable source** | Uses variable from outside | Stores variable inside |
| **Value behavior** | Value can change | Value is remembered |
| **Memory** | No memory | Has memory |
| **Dependency** | Depends on external state | Independent (captured state) |

---

## Part 2: The Loop Problem 🧠

This is where closures get confusing!

### The Confusing Example
```python
funcs = []

for i in range(3):
    def f():
        print(i)
    funcs.append(f)
```

Now call them:
```python
funcs[0]()  
funcs[1]()  
funcs[2]()
```

---

### What Do You Expect?

0
1
2

**What Actually Happens?**
2
2
2

---

## WHY???

**Because all functions are using the SAME variable `i`**

---

##  Step-by-Step Breakdown

**Loop runs:**
i = 0 → create function f  
i = 1 → create function f  
i = 2 → create function f  

**After loop ends:**
i = 2  (final value)

**Now calling functions:**

Each function asks: **"What is `i` right now?"**

👉 Answer = **2** (the current value of `i`)

So all three functions print **2**

---

##  KEY IDEA (THIS IS EVERYTHING)

### 👉 Functions remember the **variable**, not the **value**
### 👉 And the variable is **shared** → so all see the **latest value**

---

##  Solution 1: Default Argument
```python
funcs = []

for i in range(3):
    def f(n=i):  # Capture i NOW
        print(n)
    funcs.append(f)

funcs[0]()  # Output: 0
funcs[1]()  # Output: 1
funcs[2]()  # Output: 2
```

**Why This Works?**

At creation time:
i = 0 → n = 0 stored in function  
i = 1 → n = 1 stored in function  
i = 2 → n = 2 stored in function  

Now each function has **its own value** stored as a default argument.

---

## 🔥 Solution 2: Factory Function (Best for Understanding)
```python
def make_func(n):
    def f():
        print(n)
    return f

funcs = [make_func(i) for i in range(3)]

funcs[0]()  # Output: 0
funcs[1]()  # Output: 1
funcs[2]()  # Output: 2
```

**Why This Works?**

Each call to `make_func(i)` creates a **new closure** with its **own `n`**.
make_func(0) → creates closure with n=0  
make_func(1) → creates closure with n=1  
make_func(2) → creates closure with n=2  

Each closure has **independent memory**.

---

##  Real-Life Analogy

###  Problem Version (Shared Whiteboard)

Think of `i` as a **shared whiteboard**:
* All functions read from the **same board**
* Final value = **2**
* Everyone prints **2**

### ✅ Fixed Version (Individual Papers)

Each function gets its **own paper**:
* **func1** → paper has **0**
* **func2** → paper has **1**
* **func3** → paper has **2**

---

## 🧠 Final Clarity (Super Simple)

### Regular Function:
👉 **Looks at variable when running**
```python
x = 10
def func():
    print(x)  # Looks up x NOW

func()  # Output: 10
x = 20
func()  # Output: 20  (looks up again)
```

### Closure:
👉 **Remembers variable when created**
```python
def outer(x):
    def inner():
        print(x)  # Remembers x from creation
    return inner

f = outer(10)  # Captures x=10
f()  # Output: 10 (always 10, even if outer's x changes)
```

---

## Complete Comparison Table

| Aspect | Regular Function | Closure |
|--------|------------------|---------|
| **When variable is captured** | At call time (runtime) | At creation time (definition) |
| **Variable scope** | Global/module scope | Enclosing function scope |
| **Shared state** | All calls share same variable | Each closure has own copy |
| **Memory** | No extra memory | Stores captured variables |
| **Independence** | Dependent on external state | Independent after creation |


## How to Identify a Closure

Ask these questions:

1. **Is there a nested function?** ✅
2. **Does the nested function use variables from the outer function?** ✅
3. **Is the nested function returned or passed around?** ✅
