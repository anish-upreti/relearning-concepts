## Context Managers in Python

A context manager is used to manage resources automatically using the `with` statement.
A context manager is a Python feature that automatically handles setup and cleanup around a block of code using `with`.

It performs:
- setup before the block starts
- cleanup after the block ends

Example:
```python
with open("file.txt", "r") as f:
    content = f.read()

### Why do we need context managers?

Because many things in programming need:

- setup  
- use  
- cleanup  

Examples:
- opening files  
- database connections  
- locks in threading  
- network connections  
- temporary resources  

Without context managers, people may forget cleanup.

### 🧱 Basic structure

```python
with something as name:
    # do work

General meaning:

something enters a context
name is what you use inside the block
when the block ends, Python exits the context automatically

````python
# # Code example:
with open("sample.txt", "r") as f:
    content = f.read()

# # the above is equivalen to:
f = open("sample.txt", "r")
try:
    content = f.read()
finally:
    f.close()
````

### How it works internally

A context manager uses two special methods:  
  
__enter__(): Runs when the with block starts.

__exit__():Runs when the with block ends.

````python
class MyContext:
    def __enter__(self):
        print("Entering the block")
        return self
    
    def __exit__(self, exec_type, exec_value, traceback):
        print("Exiting the block")

with MyContext():
    print("Inside the block")
````

### Why return self in __enter__()?

Because whatever __enter__() returns becomes the variable after as.

````python
class MyContext:
    def __enter__(self):
        print("Start")
        return "hello"

    def __exit__(self, exc_type, exc_value, traceback):
        print("End")

with MyContext() as value:
    print(value)
````

__enter__() returned "hello"
that became value

Even though an error occurs, __exit__() still runs.

### Meaning of __exit__(self, exc_type, exc_value, traceback)

These parameters give info about any exception:  

exc_type → type of error   
exc_value → actual error object/message  
traceback → traceback details  

If no exception occurs, all are None.

### Can a context manager suppress errors?

Yes.  

If __exit__() returns True, Python treats the exception as handled.  
But we usually do not suppress errors unless needed.

````python
class IgnoreError:
    def __enter__(self):
        print("Entering")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting")
        return True

with IgnoreError():
    print(10 / 0)

print("Program continues")
````

````python
class Open_File:

    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file
    
    def __exit__(self, exc_type, exc, tb):
        self.file.close()


with Open_File("sample.txt", "w+") as f:
    f.write("this is sample text")
    f.seek(0)   # move cursor back to start before reading
    content = f.read()
    print(content)

print(f.closed)
````

````python
## Replicating context manager using decorators and generators
from contextlib import contextmanager

@contextmanager
def open_file(filename, mode):
    try:
        f = open(filename, mode)   
        yield f
    finally:     
        f.close()     

with open_file("sample.txt", "w+") as f:
    f.write("this is sample text")
    f.seek(0)   # move cursor back to start before reading
    content = f.read()
    print(content)


print(f.closed)
````
