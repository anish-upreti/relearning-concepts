# Prefix Sums

---

## What is a Prefix Sum?

A prefix sum array stores the **cumulative sum up to each index**. Instead of recomputing sums repeatedly, you precompute once and answer any range query in O(1).

```
original:  [3,  1,  4,  1,  5]
            ↓   ↓   ↓   ↓   ↓
prefix:  [0, 3,  4,  8,  9, 14]
prefix[0] = 0 — nothing added yet
prefix[1] = 3 — added first element (3)
prefix[2] = 4 — added second element (3+1)
prefix[3] = 8 — added third element (3+1+4)
and so on...
So prefix[i] answers the question: "what is the sum of all elements before index i?"
```

Each value in the prefix array = sum of all original values up to that point.

### The core formula

**Build:**
```
prefix[i] = prefix[i-1] + nums[i-1]
```

**Range query — sum from index i to j (inclusive):**
```
range_sum(i, j) = prefix[j+1] - prefix[i]

range_sum(start, end) = prefix[end+1] - prefix[start]

```

Without prefix sums, every range query loops through the range → O(n) per query.  
With prefix sums, every range query is one subtraction → O(1) per query.

---
## Part 1 — Build a Prefix Sum Array

The prefix array is always one element longer than the original.  
It starts with `0` — this makes the range query formula work cleanly for any range including index 0.

````python
# build the prefix sum array for nums
# prefix[0] = 0
# prefix[i] = prefix[i-1] + nums[i-1]

nums = [3, 1, 4, 1, 5]

prefix = [0]  # start with 0

for num in nums:
    prefix.append(prefix[-1] + num)      # last value in prefix  +  current num

print("nums:  ", nums)
print("prefix:", prefix)
# expected: [0, 3, 4, 8, 9, 14]
````

---
## Part 2 — Range Query in O(1)

Once prefix is built, any range sum is a single subtraction:

```
range_sum(i, j) = prefix[j+1] - prefix[i]
```

**Why does this work?**

```
prefix[j+1] = sum of everything from index 0 to j
prefix[i]   = sum of everything from index 0 to i-1

subtracting cancels out the part before i, leaving only i → j
```

Think of it like bank account balances:
- balance at day 10 = 500
- balance at day 3  = 200
- money earned between day 3 and day 10 = 500 - 200 = 300

````python
# implement range_sum using the prefix array you built above
# range_sum(i, j) should return the sum of nums[i] through nums[j] inclusive

def range_sum(prefix, i, j):
     return prefix[j+1] - prefix[i]

# test
print(range_sum(prefix, 0, 2))  # 3+1+4 = 8
print(range_sum(prefix, 1, 3))  # 1+4+1 = 6
print(range_sum(prefix, 2, 4))  # 4+1+5 = 10
print(range_sum(prefix, 0, 4))  # 3+1+4+1+5 = 14
````

---
## Part 3 — Why This Matters: Naive vs Prefix

Imagine you have 1 million data points (a year of hourly sensor readings) and you need to answer 10,000 range queries.

**Naive approach:** loop through the range every query
```
O(n) per query × 10,000 queries = 10,000,000,000 operations
```

**Prefix sum approach:** build once, query instantly
```
O(n) to build + O(1) per query  = 1,010,000 operations
```

The prefix array trades a one-time O(n) build cost for O(1) queries forever after. This is the same trade-off as database indexing — pay once upfront, save on every lookup.

````python
# demonstrate the difference — time both approaches on a large array
import time

large_nums = list(range(1, 100001))  # [1, 2, 3, ..., 100000]
queries = [(i, i + 1000) for i in range(0, 10000)]  # 10,000 range queries

# naive approach — loop through range every time
start = time.time()
for i, j in queries:
    total = sum(large_nums[i:j+1])
naive_time = time.time() - start

# prefix sum approach — build once, query in O(1)
# step 1: build the prefix array for large_nums
large_prefix = [0]
for  num in large_nums:
    large_prefix.append(large_prefix[-1] + num)

# step 2: answer all queries using range_sum
start = time.time()
for i, j in queries:
    total = range_sum(large_prefix, i, j)
prefix_time = time.time() - start

print(f"Naive:  {naive_time:.4f}s")
print(f"Prefix: {prefix_time:.4f}s")
print(f"Speedup: {naive_time / prefix_time:.1f}x faster")
````

## Part 4 — Real-World Use: Time Series & Feature Engineering

Prefix sums are the engine behind several ML preprocessing patterns:

| Use Case | What prefix sum does |
|----------|---------------------|
| **Rolling window features** | sum over last N days = prefix[i] - prefix[i-N] |
| **Cumulative revenue charts** | prefix sum of daily revenue |
| **Anomaly detection** | compare window sum to expected range |
| **NLP — subarray of tokens** | sum token weights in a span |

A rolling 7-day sales sum for every day in a year — with prefix sums that's O(n), without it's O(n × 7).

````python
# real-world example: compute rolling 3-day revenue sum for each day
# rolling_sum[i] = revenue[i-2] + revenue[i-1] + revenue[i]  (last 3 days)
# only compute for days where a full 3-day window exists (i >= 2)

daily_revenue = [100, 200, 150, 300, 250, 400, 180]

# step 1: build prefix array
rev_prefix = [0]
for rev in daily_revenue:
    rev_prefix.append(rev_prefix[-1] + rev)

# step 2: compute rolling 3-day sums using range_sum
# for each day i (starting at index 2), sum = range_sum(prefix, i-2, i)
rolling_sums = []
for i in range(2, len(daily_revenue)):
    result = range_sum(rev_prefix, i-2, i)
    rolling_sums.append(result)


print("Daily revenue:", daily_revenue)
print("Rolling 3-day sums:", rolling_sums)
# expected: [450, 650, 700, 950, 830]
````

````python
# PRODUCT OF ARRAY EXCEPT SELF
# given an array nums, return an array output where output[i] is the
# product of all elements EXCEPT nums[i]
# example: [1,2,3,4] → [24,12,8,6]
# constraint: O(n) time, no division operator
#
# KEY INSIGHT: output[i] = (product of everything to the LEFT of i)
#                         × (product of everything to the RIGHT of i)
#
# this is the prefix sum idea applied to multiplication instead of addition:
# build a prefix product array and a suffix product array, then multiply them
#
# prefix[i] = product of nums[0] to nums[i-1]
# suffix[i] = product of nums[i+1] to nums[n-1]
# output[i] = prefix[i] * suffix[i+1]

def product_except_self(nums):
    n = len(nums)

    # prefix product
    prefix = [1]
    for num in nums:
        prefix.append(prefix[-1] * num)

    suffix = [1]
    # suffix product - we need to loop BACKWARDS here
    for num in reversed(nums):
        suffix.append(suffix[-1] * num)
    suffix.reverse()    # flip it so suffix[i] aligs with nums[i]

    # multiply
    output = []
    for i in range(n):
        output.append(prefix[i] * suffix[i+1])
    
    return output

print(product_except_self([1, 2, 3, 4]))   # [24, 12, 8, 6]
print(product_except_self([-1, 1, 0, -3, 3]))  # [0, 0, 9, 0, 0]
````

```python
prefix is built left → right, starts with [1]
  → prefix[i] naturally aligns with "left of index i" ✓

suffix is built right → left, starts with [1], then reversed
  → after reversing, suffix[i] is shifted by 1
  → so we need suffix[i+1] to mean "right of index i"
```

3. In Product of Array Except Self, why can't you just divide the total product by nums[i]? What edge case breaks that?  
=>The division approach breaks when there's a zero in the array. If nums[i] = 0, we will be dividing total_product / 0 which is undefined and crashes. Even if we tried to handle it, we can't recover the correct answer — once the total product is 0, division can never give us back the non-zero value that belongs at that position. The prefix/suffix approach handles zeros naturally because it never divides — it just multiplies two independent sides together.  

4. Where would you use prefix sums in a real ML pipeline? Give one concrete example.  
=>In a prediction pipeline, suppose we have a table of daily user activity (clicks, purchases, time spent). For each day, we want a feature- "total activity over the last 7 days". Without prefix sums we will have to loop back 7 days for every single row — O(n * 7). With a prefix sum built once on the activity column, every 7-day window sum is just one subtraction — O(1) per row. This is exactly what pandas .rolling(7).sum() does internally, and it's a standard feature engineering step before feeding data into models like XGBoost or logistic regression.
