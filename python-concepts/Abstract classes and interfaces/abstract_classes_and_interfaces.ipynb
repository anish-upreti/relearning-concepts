### Abstract class  
Abstract class is a class that cannot be instantiated directly and may contain abstract methods as well as normal methods. It provides a common base for derived classes.

```python
from abc import ABC, abstractmethod   # abc - abstract base classes , ABC - abstract base class

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# if i try to instantiate the abstract class shape, it will throw error
shape_1 = Shape()
```

```python
from abc import ABC, abstractmethod   # abc - abstract base classes , ABC - abstract base class

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# if I inherit from abstract class then the child class is supposed to implement the abstract method
class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b
    
rect_1 = Rectangle(4, 5)
rect_1.area()
```

Here:  
Shape is abstract class  
area() is abstract method  
Rectangle must define area() 

### Interface

An interface is a contract that defines what a class must do.

It specifies a set of methods that a class must implement, but it does not define how those methods should work.

### Key Idea

An interface describes behavior (ability), not identity.

“If a class uses this interface, it MUST implement these methods.”

### Example Concept

If a class implements a Flyable interface, it must have:

fly()

Different classes can implement it in their own way:

Bird → fly using wings
Airplane → fly using engine

They are not related, but share the same behavior.

### Python Implementation

Python does not have a separate interface keyword.  
We use abstract classes (ABC) to create interface-like behavior.

```python
from abc import ABC, abstractmethod

class Flyable(ABC):   # acting like interface

    @abstractmethod
    def fly(self):
        pass


class Bird(Flyable):
    def fly(self):
        return "Bird is flying"
    
class Aeroplane(Flyable):
    def fly(self):
        return "Aeroplane is flying"
    
bird = Bird()
print(bird.fly())

# same as
# Bird.fly(bird)

aeroplane = Aeroplane()
print(Aeroplane.fly(aeroplane))
```

### 💡 Key Insight  
#### Abstract Class → WHAT you are  
Animal  
Vehicle  
Shape   
#### Interface → WHAT you can do  
Flyable  
Drivable  
Printable  

###  Why Interfaces Exist

Because sometimes:  

classes are unrelated  
but behavior is same  

So we need a common contract, not a common parent.

| Abstract Class                            | Interface                            |
| ----------------------------------------- | ------------------------------------ |
| Represents **what an object is**          | Represents **what an object can do** |
| Defines common base structure             | Defines common behavior              |
| Can have both abstract and normal methods | Usually only abstract methods        |
| Can have instance variables               | Usually no instance variables        |
| Used for closely related classes          | Used for unrelated classes           |
| Uses inheritance (`is-a`)                 | Uses capability (`can-do`)           |


### Simple Understanding  
Abstract Class → “is-a” relationship
Example: Dog is an Animal  
### Interface → “can-do” relationship   
Example: Bird can fly, Airplane can fly
### One-Line Summary. 
Abstract Class = Blueprint (identity)
Interface = Contract (ability)

```python
# 1. Create an abstract base class: Shape
#    abstract methods:
#    - area() — must be implemented by subclasses
#    - perimeter() — must be implemented by subclasses
#    - @property description — must be implemented by child class(@abstract method)
#    concrete method (same for all shapes):
#    - compare_area(other) — returns which shape
#      has bigger area, e.g. "Circle is bigger"

# 2. Circle — inherits from Shape
#    - __init__: radius
#    - implement area() → π * r²
#    - implement perimeter() → 2 * π * r
#    - implement description → "Circle with radius X"

# 3. Rectangle — inherits from Shape
#    - __init__: width, height
#    - implement area() → width * height
#    - implement perimeter() → 2 * (width + height)
#    - implement description → "Rectangle WxH"

# 4. Triangle — inherits from Shape
#    - __init__: base, height, side1, side2, side3
#    - implement area() → 0.5 * base * height
#    - implement perimeter() → side1+side2+side3
#    - implement description → "Triangle base X"

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    @property
    @abstractmethod
    def description(self):
        pass
    
    def compare_area(self, other):
        if self.area() > other.area():
            return "{} is bigger".format(self.__class__.__name__)    # can also use type(self).__name__
        elif self.area == other.area:
            return "both have same area"
        else:
            return "{} is bigger".format(other.__class__.__name__)   # can also use type(other).__name__
        
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        area = (math.pi * self.radius * self.radius)
        return area
    
    def perimeter(self):
        return 2 * math.pi * self.radius
    
    @property
    def description(self):
        return "Circle with radius {}".format(self.radius)
    
class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b
    
    def perimeter(self):
        return 2 * (self.l + self.b)

    @property
    def description(self):
        return "Rectangle {}X{}".format(self.l, self.b)
    
class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3):
        self.base = base
        self.height = height
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def area(self):
        return 0.5 * self.base * self.height
    
    def perimeter(self):
        return self.side1 + self.side2 + self.side3

    @property
    def description(self):
        return f"Triangle base {self.base}"
    
```

```python
# test
c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(3, 4, 3, 4, 5)

print(c.area())            # 78.53981633974483
print(r.perimeter())       # 20
print(c.compare_area(r))   # Circle is bigger
print(t.description)       # Triangle base 3

```

add_book() as an abstract method  
remove_book() as an abstract method   
find_book() as a concrete method (same logic for all libraries)  

Then make DigitalLibrary inherit from it and implement those abstract methods.

```python
from abc import ABC, abstractmethod

class Book:

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author} ({self.pages} pages)"
    
    def __repr__(self):
        return "Book({}, {}, {})".format(self.title, self.author, self.pages)
    
    @property
    def summary(self):
        return "Title: {} | Author: {} | Pages: {}".format(self.title, self.author, self.pages)
    

class Library(ABC):
    total_libraries = 0
    def __init__(self, name, city):
        self.name = name
        self.city = city
        self.books = []
        Library.total_libraries += 1

    @abstractmethod
    def add_book(self, book):
        pass

    @abstractmethod
    def remove_book(self, title):
        pass
    
    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                return book
        return None
    
    def __len__(self):
        return len(self.books)
    
    def __str__(self):
        return "{} Library - {} books".format(self.name, len(self.books))
    

class DigitalLibrary(Library):
    max_books = 5

    def __init__(self, name, city):
        super().__init__(name, city)

    def add_book(self, book):
        if len(self.books) >= self.max_books:
            print("Library full")
        else:
            self.books.append(book)

    def remove_book(self, title):
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                return
        print("Book not found")

    @classmethod
    def from_string(cls, s):
        name, city = s.split("-")
        return cls(name, city)
```
