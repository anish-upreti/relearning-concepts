### Class variables
It is shared across all instances of that class.
It can be accessed with help of class itself or instance of that class.

### Difference Between Class Variable and Instance Variable

A **class variable** is shared by all objects (instances) of a class. It is defined inside the class but outside any method. This means every instance sees the same value unless it is specifically overridden.

Example:
- `interest_rate = 0.05` → all bank accounts use the same interest rate.

An **instance variable** is unique to each object. It is defined inside the `__init__` method using `self`. Each instance can have different values.

Example:
- `self.balance` → each account has its own balance.

### Key Difference:
- Class variable → **shared across all objects**
- Instance variable → **unique per object**

### Simple Analogy:
Think of a class as a template for bank accounts:
- Class variable → like a **common rule** (same interest rate for all accounts)
- Instance variable → like **personal data** (each person’s account balance)

````python
class Employee:
    
    # class variable
    num_of_employees = 0
    raise_rate = 1.04

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay

        # accessing class variable with help of class itself
        Employee.num_of_employees += 1

    def apply_raise(self):
        self.pay = self.pay * self.raise_rate # accessing raise rate class variable through instance

````

````python
emp1 = Employee("anish", "upreti", 100000)
print(emp1.pay)

emp1.apply_raise()

print(emp1.pay)

print(emp1.raise_rate)
print(Employee.raise_rate)
print(Employee.num_of_employees)
````

````python
# Add these to your existing BankAccount class:

# 1. A class variable — interest_rate = 0.05
#    (shared across ALL accounts, not per instance)

# 2. A method — apply_interest(self)
#    adds interest to the current balance

# 3. A class method — set_interest_rate(cls, rate)
#    changes the rate for ALL accounts at once

class BankAccount:
    
    # 1. Class variable (shared by ALL instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables (unique for each object)
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return("Insufficient funds")
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object (used in print)
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # 2. Add interest to the current balance
        # Uses class variable (interest_rate)
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # 3. Class method to update interest rate for ALL accounts
        # cls refers to the class (BankAccount)
        cls.interest_rate = rate


# Create two account objects
acc1 = BankAccount("ram", 1000)
acc2 = BankAccount("hari", 2000)

# Apply interest using default rate
acc1.apply_interest()
print(acc1.get_balance())

# Change interest rate for ALL accounts
BankAccount.set_interest_rate(0.10)

# Apply new interest rate
acc1.apply_interest()
print(acc1.get_balance()) 

acc2.apply_interest()
print(acc2.get_balance())   

acc2.apply_interest()
print(acc2.get_balance())
````

### Class methods and static methods
Regular methods in a class automatically take the instance as the first argument.  
To take class as first argument we use class methods.(use @classmethod)  
Static methods do not pass anything automatically unlike regular and class methods. (use @staticmethod)


````python
class Employee:
    
    # class variable
    num_of_employees = 0
    raise_rate = 1.04

    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay

        # accessing class variable with help of class itself
        Employee.num_of_employees += 1

    def apply_raise(self):
        self.pay = self.pay * self.raise_rate # accessing raise rate class variable through instance

    # using class method lets us use the class variable in the function
    @classmethod
    def from_string(cls, emp_str):
        first, last, pay = emp_str.split("-")
        return cls(first, last, pay)
    
    # using static method lets us use the function without passing in instance or class argument
    def is_workday(day):
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True
````

### When to use classmethod vs staticmethod vs regular method

A **regular method** (uses `self`) is used when you need to work with instance data, like updating a specific account’s balance.  

A **classmethod** (uses `cls`) is used when you want to work with or modify class-level data that affects all instances, like changing the interest rate.  

A **staticmethod** is used when the method doesn’t need access to either instance or class data, and is just a utility function related to the class.

### Q. Solve the below problem.


````python
# Add to your existing BankAccount class:

# 1. A classmethod — from_string(cls, account_str)
#    takes a string like "Alice-1000" and creates
#    a BankAccount instance from it

# 2. A classmethod — from_dict(cls, data)
#    takes a dict like {"owner": "Bob", "balance": 500}
#    and creates a BankAccount instance

# 3. A staticmethod — validate_amount(amount)
#    returns True if amount > 0, False otherwise
#    use this inside deposit() and withdraw()
#    to validate before doing anything

class BankAccount:
    
    # Class variable (shared by all instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        # Use staticmethod to validate amount first
        if BankAccount.validate_amount(amount):
            self.balance += amount
            return self.balance
        return "Invalid amount"

    def withdraw(self, amount):
        # Validate amount first
        if not BankAccount.validate_amount(amount):
            return "Invalid amount"
        
        # Then check if enough balance is available
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return "Insufficient funds"
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # Adds interest to the current balance
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # Changes interest rate for all accounts
        cls.interest_rate = rate

    @classmethod
    def from_string(cls, acc_str):
        # Create object from string like "Alice-1000"
        name, balance = acc_str.split('-')
        return cls(name, float(balance))
    
    @classmethod
    def from_dict(cls, acc_dict):
        # Create object from dictionary like {"owner": "Bob", "balance": 500}
        name = acc_dict["owner"]
        balance = acc_dict["balance"]
        return cls(name, balance)
    
    @staticmethod
    def validate_amount(amount):
        # Returns True only if amount is greater than 0
        return amount > 0
    
````

````python
# testing the above class
acc1 = BankAccount.from_string("Alice-1000")
acc2 = BankAccount.from_dict({"owner": "Bob", "balance": 500})
print(acc1)
print(acc2)
print(BankAccount.validate_amount(-50))  # False
print(BankAccount.validate_amount(200))
````

### Inheritance - Creating subclasses


````python
class Employee:
    
    # Class variable shared by all employees
    raise_amt = 1.04

    # Constructor / initializer
    def __init__(self, first, last, pay):
        # Instance variables (different for each object)
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'

    def fullname(self):
        # Returns full name of employee
        return '{} {}'.format(self.first, self.last)
    
    def apply_raise(self):
        # Updates pay using class variable raise_amt
        self.pay = int(self.pay * self.raise_amt)


# creating subclass
class Developer(Employee):
    pass
````

````python
# Creating subclass of Employee
class Developer(Employee):
    
    # Developer has its own raise amount
    raise_amt = 1.15

    def __init__(self, first, last, pay, prog_lang):
        # Call parent class constructor to reuse first, last, pay setup
        super().__init__(first, last, pay)

        # Extra attribute only for Developer
        self.prog_lang = prog_lang

dev_2 = Developer("tony", "stark", 10000, "python")
dev_2.apply_raise()
print(dev_2.pay)

dev_3 = Developer("Captain", "america", 1000, "python")
print(dev_3.prog_lang)
````

````python
# Another subclass of Employee
class Manager(Employee):
    
    def __init__(self, first, last, pay, employees=None):
        # Initialize inherited attributes from Employee
        super().__init__(first, last, pay)

        # Avoid using mutable default argument like employees=[]
        if employees is None:
            self.employees = []
        else:
            self.employees = employees

    def add_emp(self, emp):
        # Add employee only if not already in the list
        if emp not in self.employees:
            self.employees.append(emp)

    def remove_emp(self, emp):
        # Remove employee if present in the list
        if emp in self.employees:
            self.employees.remove(emp)
        
    def print_emps(self):
        # Print full names of all employees managed by this manager
        for emp in self.employees:
            print("-->", emp.fullname())


````

````python
manager_1.remove_emp(dev_2)
manager_1.print_emps()
````

### super() vs direct class call (complete note with examples)

## Key Idea
- `super()` does **not** mean “go to parent”
- It means “go to the **next class in MRO** (Method Resolution Order)”
- Direct call like `A.__init__(self)` or `Employee.__init__(self, ...)` means:
  - “jump directly to this exact class”

---

## 1) Example using `super()` in multiple inheritance

```python
class A:
    def __init__(self):
        print("A")

class B(A):
    def __init__(self):
        print("B start")
        super().__init__()
        print("B end")

class C(A):
    def __init__(self):
        print("C start")
        super().__init__()
        print("C end")

class D(B, C):
    def __init__(self):
        print("D start")
        super().__init__()
        print("D end")

d = D()
Output
D start
B start
C start
A
C end
B end
D end
Why?

Python follows the MRO:

D → B → C → A

So:

D starts
D calls next → B
B calls next → C
C calls next → A
then control returns back:
C end
B end
D end
Important understanding

Each method:

runs code before super()
calls super()
pauses
waits for next class to finish
resumes and runs remaining code

So:

going down the chain = start prints
coming back up = end prints
2) Example using direct call like A.__init__(self)
class A:
    def __init__(self):
        print("A")

class B(A):
    def __init__(self):
        print("B start")
        A.__init__(self)   # direct call
        print("B end")

class C(A):
    def __init__(self):
        print("C start")
        A.__init__(self)   # direct call
        print("C end")

class D(B, C):
    def __init__(self):
        print("D start")
        B.__init__(self)   # direct call
        C.__init__(self)   # direct call
        print("D end")

d = D()
Output
D start
B start
A
B end
C start
A
C end
D end
What problem happened?

A got called two times.

Why?
Because:

B directly jumped to A
C also directly jumped to A

So direct call can:

duplicate calls
break the inheritance chain
make code harder to maintain

### Solve the below question


````python
# 1. SavingsAccount — inherits from BankAccount
#    - class variable: withdrawal_limit = 3
#      (can only withdraw 3 times, then blocked)
#    - override withdraw() to enforce this limit
#    - add a method: reset_limit()
#      resets the withdrawal count back to 0

# 2. BusinessAccount — inherits from BankAccount
#    - class variable: transaction_fee = 10
#      (every withdrawal charges a $10 fee)
#    - override withdraw() to deduct fee on top
#    - override __str__ to show "Business Account"

class BankAccount:
    
    # Class variable (shared by all instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        # Use staticmethod to validate amount first
        if BankAccount.validate_amount(amount):
            self.balance += amount
            return self.balance
        return "Invalid amount"

    def withdraw(self, amount):
        # Validate amount first
        if not BankAccount.validate_amount(amount):
            return "Invalid amount"
        
        # Then check if enough balance is available
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return "Insufficient funds"
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # Adds interest to the current balance
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # Changes interest rate for all accounts
        cls.interest_rate = rate

    @classmethod
    def from_string(cls, acc_str):
        # Create object from string like "Alice-1000"
        name, balance = acc_str.split('-')
        return cls(name, float(balance))
    
    @classmethod
    def from_dict(cls, acc_dict):
        # Create object from dictionary like {"owner": "Bob", "balance": 500}
        name = acc_dict["owner"]
        balance = acc_dict["balance"]
        return cls(name, balance)
    
    @staticmethod
    def validate_amount(amount):
        # Returns True only if amount is greater than 0
        return amount > 0


class SavingsAccount(BankAccount):
    withdrawal_limit = 3

    def __init__(self, name, balance):
        super().__init__(name, balance)
        self.withdraw_count = 0

    def withdraw(self, withdraw_amt):
        # Check if withdrawal limit is reached FIRST
        if self.withdraw_count >= self.withdrawal_limit:
            print("Withdrawal limit reached")
            return "Limit reached"

        # Then call parent's withdraw method
        result = super().withdraw(withdraw_amt)

        # Only increment count if withdrawal was successful
        if result != "Insufficient funds" and result != "Invalid amount":
            self.withdraw_count += 1
            print(f"${withdraw_amt} withdrawn successfully. Balance: ${result}")
            return result
        
        return result
        
    def reset_limit(self):
        # Reset withdrawal count back to 0
        self.withdraw_count = 0
        return "Withdrawal count reset"


class BusinessAccount(BankAccount):
    transaction_fee = 10

    def __init__(self, name, balance):
        super().__init__(name, balance)

    def withdraw(self, withdraw_amt):
        # Calculate total amount including fee
        total_amount = withdraw_amt + self.transaction_fee
        
        # Call parent's withdraw with total amount
        result = super().withdraw(total_amount)
        
        # Provide feedback
        if result != "Insufficient funds" and result != "Invalid amount":
            print(f"${withdraw_amt} withdrawn + ${self.transaction_fee} fee = ${total_amount} total. Balance: ${result}")
        
        return result

    def __str__(self):
        # Override to show "Business Account"
        return f"Business Account[{self.name}] — Balance: ${self.balance}"
````

````python
# Test it:
sav = SavingsAccount("Alice", 1000)
sav.withdraw(100)  # works
sav.withdraw(100)  # works
sav.withdraw(100)  # works
sav.withdraw(100)  # blocked — limit reached
print(sav.reset_limit())

biz = BusinessAccount("Bob", 2000)
biz.withdraw(500)  # deducts 510 total (500 + fee)
print(biz)
````

````python
# Add these dunder methods to BankAccount:

# 1. __str__      — "Account[Alice] — Balance: $1000"
#                   (readable, for users)

# 2. __repr__     — "BankAccount('Alice', 1000)"
#                   (unambiguous, for developers/debugging)

# 3. __len__      — returns the balance as an integer
#                   so len(acc1) gives the balance

# 4. __eq__       — two accounts are equal if they
#                   have the same owner AND balance

# 5. __add__      — acc1 + acc2 creates a NEW account
#                   owner = "Alice & Bob"
#                   balance = combined balance of both


class BankAccount:
    
    # Class variable (shared by all instances)
    interest_rate = 0.05

    def __init__(self, name, balance):
        # Instance variables
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        # Use staticmethod to validate amount first
        if BankAccount.validate_amount(amount):
            self.balance += amount
            return self.balance
        return "Invalid amount"

    def withdraw(self, amount):
        # Validate amount first
        if not BankAccount.validate_amount(amount):
            return "Invalid amount"
        
        # Then check if enough balance is available
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return "Insufficient funds"
    
    def get_balance(self):
        # Returns current balance
        return self.balance

    def __str__(self):
        # String representation of object
        return f"Account[{self.name}] — Balance: ${self.balance}"
    
    def apply_interest(self):
        # Adds interest to the current balance
        self.balance = self.balance + (self.balance * self.interest_rate)

    @classmethod
    def set_interest_rate(cls, rate):
        # Changes interest rate for all accounts
        cls.interest_rate = rate

    @classmethod
    def from_string(cls, acc_str):
        # Create object from string like "Alice-1000"
        name, balance = acc_str.split('-')
        return cls(name, float(balance))
    
    @classmethod
    def from_dict(cls, acc_dict):
        # Create object from dictionary like {"owner": "Bob", "balance": 500}
        name = acc_dict["owner"]
        balance = acc_dict["balance"]
        return cls(name, balance)
    
    @staticmethod
    def validate_amount(amount):
        # Returns True only if amount is greater than 0
        return amount > 0
    
    def __repr__(self):
        return f"BankAccount('{self.name}', {self.balance})"

    def __len__(self):
        return int(self.balance)
    
    def __eq__(self, other):
        if self.name == other.name and self.balance == other.balance:
            return True
        else:
            return False
        
    # def __eq__(self, other):
    #     # Two accounts are equal if name and balance are same
    #     return self.name == other.name and self.balance == other.balance

    def __add__(self, other):
        # acc1 + acc2 creates a NEW BankAccount object
        new_name = f"{self.name} & {other.name}"
        new_balance = self.balance + other.balance
        return BankAccount(new_name, new_balance)
````

### Property Decorators
Property Decorators in Python

Property decorators are used to control access to class attributes in a clean and Pythonic way. They let us define methods that act like attributes.

The main property decorators are:

1. @property
Used to make a method behave like a readable attribute.
It is called a getter.

2. @<property_name>.setter
Used to set or update the value of that property.
It is called a setter.

3. @<property_name>.deleter
Used to delete a property value.
It is called a deleter.

````python
class Employee:
    
    raise_amt = 1.04

    def __init__(self, first, last):
        self.first = first
        self.last = last

    # Getter for email
    @property
    def email(self):
        return self.first + '.' + self.last + '@company.com'

    # Getter for fullname
    @property
    def fullname(self):
        return '{} {}'.format(self.first, self.last)

    # Setter for fullname
    @fullname.setter
    def fullname(self, name):
        first, last = name.split(' ')
        self.first = first
        self.last = last

    # Deleter
    @fullname.deleter
    def fullname(self):
        print("fullname deleted")
        self.first = None
        self.last = None
````

````python
emp_1 = Employee("Ram", "Hari")
emp_1.first = "Shyam"
print(emp_1.email)
print(emp_1.fullname)
````

````python
emp_1.fullname = "Hari Prasad"
print(emp_1.first)
print(emp_1.last)
````

````python
del emp_1.fullname
print(emp_1.first)
````

````python
# Refactor BankAccount using property decorators:

# 1. Make balance a private variable: _balance
#    (rename self.balance to self._balance everywhere)

# 2. @property — balance
#    getter that returns self._balance

# 3. @balance.setter — balance
#    validates before setting:
#    - must be a number
#    - cannot be negative
#    - raise ValueError if invalid

# 4. @property — email
#    auto-generates email from owner name
#    e.g. "alice@bank.com" (lowercase, spaces → no spaces)

# 5. @email.setter
#    raise AttributeError("Cannot set email directly")

class BankAccount:
    def __init__(self, fullname, balance):
        self.fullname = fullname
        self._balance = balance  #private variable

    #getter
    @property
    def balance(self):
        return self._balance
    
    #setter with validation
    @balance.setter
    def balance(self, value):
        if not isinstance(value, (int, float)):
            raise ValueError("Balance must be a number")
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self._balance = value

    # email generation from name
    @property
    def email(self):
        clean_name = self.fullname.lower().replace(" ", "")
        return f"{clean_name}@bank.com"
    
    @email.setter
    def email(self, value):
        raise AttributeError("Cannot set email directly")
    
    
````

````python
# Test it:
acc1 = BankAccount("Alice Smith", 1000)
print(acc1.balance)       # 1000
print(acc1.email)         # alicesmith@bank.com
acc1.balance = 2000       # works fine
````

## Why use @property instead of public variables?

Using @property helps control how attributes are accessed and modified.

If we make variables public (like self.balance), anyone can change them directly:
    acc.balance = -1000   ❌ (invalid but allowed)

With @property:
- We can add validation (e.g., prevent negative balance)
- We can make attributes read-only (like email)
- We can compute values dynamically (email from owner name)
- We can change internal logic without changing how the attribute is used

So, @property provides:
- Encapsulation
- Data validation
- Cleaner and safer code

It lets us use methods like attributes while keeping control.
