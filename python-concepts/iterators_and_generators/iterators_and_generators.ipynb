## Iterable (Short Note)

An **iterable** is any object that can be looped over (iterated).

It means:
> The object can return its elements one by one.

---

### Examples of Iterables
- List → `[1, 2, 3]`
- Tuple → `(1, 2, 3)`
- String → `"hello"`
- Dictionary → `{"a": 1, "b": 2}`
- Set → `{1, 2, 3}`

---

### How It Works

An iterable can be converted into an iterator using:

```python
iter(obj)
One-line memory trick

👉 Iterable = collection
👉 Iterator = value provider

````python
nums = [1, 2, 3, 4]   # iteratable

# to see the list of dunder methods available
print(dir(nums))
````

````python
# to make the iterable iterator we use
i_nums = nums.__iter__()    # or its equivalent is iter(nums), it calls the iter dunder method in background
i_nums
````

````python
# Iterator is an object with a state that remembers where it is during iteration
print(next(i_nums))
print(next((i_nums)))
print(next((i_nums)))
print(next((i_nums)))
print(next((i_nums)))  # will give StopIteration
````

````python
# the for loop handles the StopIteration internally so it doesn't throw error like above
## Custom iterator
class MyRange:
    def __init__(self, first, last):
        self.value = first
        self.last = last

    def __iter__(self):
        return self

    def __next__(self):
        if self.value > self.last:
            raise StopIteration
        current = self.value
        self.value += 1
        return current
    
nums = MyRange(1, 10)

for num in nums:
    print(num)
````

## How Iteration Works (Step-by-Step)

Consider:
```python
nums = MyRange(1, 3)

for num in nums:
    print(num)
```

### Step 1 — Object Creation
```python
nums = MyRange(1, 3)
```

* `nums` is an object of class `MyRange` 
* Internal state: 
  * `value = 1` 
  * `last = 3` 

### Step 2 — `for` loop starts
```python
for num in nums:
```

Python internally does:
```python
it = iter(nums)
```

This calls:
```python
nums.__iter__()
```

Since `__iter__()` returns `self`, 👉 the object itself (`nums`) acts as the iterator.

### Step 3 — First iteration

Python calls:
```python
next(it)
```

Inside `__next__()`:
* current value = 1 
* returns `1` 
* updates `value` → 2 

So:
```python
num = 1
```

### Step 4 — Next iterations

Python keeps calling `next(it)`:
* returns 2 → `num = 2` 
* returns 3 → `num = 3` 

Each time:
* value is returned 
* internal state is updated 

### Step 5 — End of iteration

When `value > last`:
```python
raise StopIteration
```

Python catches this automatically and stops the loop.

### What is `num`?

* `num` is **not** the object 
* `num` is the **value returned** by `__next__()` 

In this example:
```python
return current
```

So `num` becomes:
* `1` → first iteration 
* `2` → second iteration 
* `3` → third iteration 

👉 Therefore, `num` is an `int`

### Key Understanding

* `nums` → iterator object (controls iteration) 
* `num` → value produced at each step

````python
# PART 1 — Custom Iterator

# Build a class: CountDown
# - __init__: takes a start number
# - __iter__: returns self
# - __next__: counts down to 0,
#   raises StopIteration when done

class CountDown:
    def __init__(self, start):
        self.value = start

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.value < 0:
            raise StopIteration
        current = self.value
        self.value -= 1
        return current
    
````

````python
# Test:
cd = CountDown(5)
for n in cd:
    print(n) 
````

## Generator

A **generator** is a special type of function that returns values one at a time using `yield`.

Instead of returning all values at once, it produces values **on demand (lazy evaluation)**.

---

A generator is an easier way to create an iterator.  

Instead of writing __iter__ and __next__, you just use:  

👉 yield

### Example

```python
def count_up(n):
    for i in range(1, n+1):
        yield i

## Behavior of `yield`

`yield` is used in a function to produce values one at a time while pausing execution.

---

### How it behaves

- **Returns a value** to the caller
- **Pauses** the function at that point
- **Saves** the current state (variables, position in code)
- **Resumes** from the same point on the next call

---

### Example
```python
def gen():
    print("Start")
    yield 1
    print("Middle")
    yield 2
    print("End")
```

**Execution:**
```python
g = gen()

next(g)  # Prints "Start", returns 1, pauses
next(g)  # Prints "Middle", returns 2, pauses
next(g)  # Prints "End", raises StopIteration
```

**Step-by-step:**
* **First call** → prints "Start", returns `1`, pauses after first `yield`
* **Second call** → resumes, prints "Middle", returns `2`, pauses after second `yield`
* **Third call** → resumes, prints "End", function ends → raises `StopIteration`

---

### Key Idea

`yield` does **not** stop the function like `return`.  
It **pauses** and **continues later** from where it left off.

---

### Comparison: `return` vs `yield`

| Feature | `return` | `yield` |
|---------|----------|---------|
| Returns value | ✅ Yes | ✅ Yes |
| Ends function | ✅ Yes | ❌ No (pauses) |
| Can be called again | ❌ No | ✅ Yes (resumes) |
| State preservation | ❌ Lost | ✅ Saved |

---

### One-Line Summary

**`yield`** = return a value **+** pause execution **+** resume later

---

### Real-World Analogy

Think of `yield` like a **bookmark**:
- You read to a certain point (yield a value)
- You place a bookmark (pause and save your place)
- Later, you open the book at the bookmark (resume from where you left off)

`return` is like **closing the book** — you can't continue from where you were.

````python
def square_numbers(nums):
    for n in nums:
        yield (n * n)

my_nums = square_numbers([1, 2, 3, 4, 5])
print(my_nums)  # prints out the generator object rather than the list, because generators don't hold the entire result in memory, it yields one result at a time
````

````python
# to actually print the list items one by one
print(next(my_nums))
print(next(my_nums))
````

````python
# can iterate the whole generator with loop, and the reason it starts printing 9 is because generators also remember the state they are in iteration
for num in my_nums:
    print(num)
````

````python
# 1. Write a generator function: `counter(start, end)`
# It should yield numbers from `start` to `end`

def counter(start, end):
    while start <= end:
        yield start
        start += 1

# Test for counter
counter_gen = counter(10, 15)
print(next(counter_gen))   # 10
print(next(counter_gen))   # 11
print(next(counter_gen))   # 12

for n in counter_gen:
    print(n)
````

````python
## Connect to your Library system
# Refactor it to be iterable:
# add __iter__ and __next__ so you can do:
## for book in lib:
##   print(book)  

class Book:

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author} ({self.pages} pages)"
    
    def __repr__(self):
        return "Book({}, {}, {})".format(self.title, self.author, self.pages)
    
    @property
    def summary(self):
        return "Title: {} | Author: {} | Pages: {}".format(self.title, self.author, self.pages)
    
class Library:
    total_libraries = 0
    def __init__(self, name, city):
        self.name = name
        self.city = city
        self.books = []
        Library.total_libraries += 1

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, title):
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                return
        print("Book not found")

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                return book
        return None
    
    def __len__(self):
        return len(self.books)
    
    def __str__(self):
        return "{} Library - {} books".format(self.name, len(self.books))
    
    def __iter__(self):
        self._index = 0
        return self
    
    def __next__(self):
        if self._index >= len(self.books):
            raise StopIteration
        book = self.books[self._index]
        self._index += 1
        return book
    
````

````python
b1 = Book("The Hobbit", "Tolkien", 300)
b2 = Book("Clean Code", "Martin", 464)
b3 = Book("Python Crash Course", "Matthes", 544)

lib = Library("City", "Kathmandu")
lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b3)

for book in lib:
    print(book)
````
